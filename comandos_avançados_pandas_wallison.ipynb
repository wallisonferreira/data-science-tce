criar uma coluna nivel_governabilidade que categoriza os municiopios em três classes:


*   A -> indice_governanca >= 0.7
*   B -> indice_governanca entre 0.4 e 0.7
*   C -> os demais

criar uma funcao e utilizar o método apply:
 
df['coluna nivel_governabilidade'] = df['indice_governanca'].apply(funcao_classe_ig)



```python
def funcao_classe_ig(ig):
  if(ig>=0.7):
    return "A"
  elif (ig>=0.4):
    return "B"
  else:
    return "C"

df['nivel_governabilidade'] = df['indice_governanca'].apply(funcao_classe_ig)
```
