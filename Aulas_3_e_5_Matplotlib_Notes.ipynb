```python
total = df.regiao.value_counts().values.sum()
total
```

```python
df.regiao.value_counts()
```

```python
def build_label(value_perc):
  value = value_perc*total/100
  return "%d (%.2f%%)" % (value, value_perc)
```

```python
plt.pie(explode=[0.2, 0, 0, 0.4, 0], autopct=build_label, colors=["#ff6961", "#77dd77", "#fdfd96", "#84b6f4", "#fdcae1"], x=df.regiao.value_counts().values, labels=df.regiao.value_counts().index, shadow=True)
plt.title("Parcela de Municípios por Região Brasileira")
plt.show()
```

## Boxplot

Documentação: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
df_filtrado = df[["regiao", "exp_vida"]]
df_filtrado.head()
```

```python
df_filtrado.boxplot(column='exp_vida', by = 'regiao')
plt.show()
```

```python
regiao_list = df_filtrado.regiao.value_counts().index.tolist()
regiao_list
```

```python
data = []
for regiao in regiao_list:
  data.append(df_filtrado[df_filtrado["regiao"] == regiao]["exp_vida"].tolist())
```

```python
plt.figure(figsize=(10,7))
plt.boxplot(x=data, labels=regiao_list,vert= True)
plt.show()
```

**IMPORTANTE!** Ao utilizar plots diretamente a partir do matplotlib, valores nulos podem ocasionar comportamentos inesperados.

```python
df_filtrado["exp_vida"].isnull().any()
```

```python
dados = []
for regiao in regiao_list:
  dados.append(df_filtrado[(df_filtrado["regiao"] == regiao) & (~df_filtrado["exp_vida"].isna())]["exp_vida"].tolist())
```

```python
pd.DataFrame(dados)
```

```python
plt.figure(figsize=(10,7))
plt.boxplot(x=dados, labels=regiao_list, vert= True)
plt.show()
```

### Somente o size_values está normalizado, mas não interfere nas cores. Somente a coluna 'Precipitação (mm)' controla as cores no gráfico.

```python
plt.figure(figsize=(9,4)) # dfinindo tamanho do gráfico
plt.scatter(x="Temperatura Media (C)", y="Consumo de cerveja (litros)", vmin=min_prec, vmax=max_prec, cmap=cm, c=df_cerveja_mes["Precipitacao (mm)"], marker="v", s=size_values, data=df_cerveja_mes)
plt.colorbar()
plt.xlabel("Temperatura Media (C)") # titulo do eixo X
plt.ylabel("Consumo de Cerveja (L)") # titulo do eixo Y
plt.title("Consumo de Cerveja em L por Data de Ocorrência") # titulo do grafico
plt.show()
```
