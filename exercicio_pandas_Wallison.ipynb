8- Há registros nulos?? (SIM). Faça as seguintes correções nos dados


*   Se não tem informações do Gender (Sexo -masculino ou feminino), exclua o registro
*   Se não tem informações de "Senior Management", atribua o valor False;
*   Se não tem informações de "Team", atribua o valor "Sales"



```python
# Verificando onde tem valores nulos
dados.isnull().any()
```

```python
# Excluindo onde Gender é nulo
dados.dropna(subset = ['Gender'], axis = 0, inplace=True)
```

```python
# Atribuindo valor False onde Senior Management é nulo
dados['Senior Management'].fillna(False, inplace=True)
```

```python
# Atribuindo valor Sales onde Team é nulo
dados['Team'].fillna('Sales', inplace=True)
```

```python
# Atribuindo valor Unknown onde First Name é nulo
dados['First Name'].fillna('Unknown', inplace=True)
```

```python
dados.isnull().any()
```

```python
import seaborn as sns
sns.set_theme()
```

### Distribuição de Gênero

```python
female = dados[dados['Gender'] == 'Female']
male = dados[dados['Gender'] == 'Male']
```

```python
dados.groupby(['Gender']).sum()
```

```python
dados.groupby(['Gender']).sum().plot(kind='pie', y='Salary', autopct='%1.0f%%');
```

```python
sns.displot(
    female['Salary'], 
    kde=True, 
    stat="density", 
    kde_kws=dict(cut=3)
);
```

```python
sns.displot(
    male['Salary'], 
    kde=True, 
    stat="density", 
    kde_kws=dict(cut=3)
)
```
